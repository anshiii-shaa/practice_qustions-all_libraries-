```python
# 13

data={"Students":["John","Emily","Anna","Micheal","Sara"],
      "Marks":[50,60,55,45,40],
      "Grade":["B","A","A","C","C"]

      }

df=pd.DataFrame(data)
Passed=[]

for i in df.index:
  if df.loc[i,"Marks"]>50:

    Passed.append("yes")
  else:
    Passed.append("No")

df["Passed"]=Passed
print(df)
```
