```python
# importing necessary libraries
import numpy as np
import pandas as pd

from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# example dataset: actual vs Predicted values
actual=np.array([3,-0.5,2,7])
predicted=np.array([2.5,0.0,2,9])
```

```python
# creating a DataFrame
df=pd.DataFrame({
    "actual": actual,
    "predicted": predicted
})
print(df)
```

# **calculating errors**

  *   mae=mean_absolute_error
  *   mse=mean_squared_error

*   **always use mae**







```python
# always use mae!!

# calcualting mean absolute error (MAE) using the function
mae= mean_absolute_error(actual,predicted)
print("mean absolute error (MAE):",mae)
```

```python
# calculating mean squared error (MSE) using the function
mse=mean_squared_error(actual,predicted)
print("mean absolute error:",mse)
```
