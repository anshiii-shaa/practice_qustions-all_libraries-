```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Plot a simple line graph where the x-values range from 0 to 10 and the y- values are the squares of the x-values. Label the x-axis as 'x' and the y-axis as 'y = x^2'. Give the plot a title of 'Simple Line Plot'.




x=np.arange(11)
y=x**2
plt.plot(x,y,"c")
plt.title("Simple LIne Plot")
plt.xlabel("X")
plt.ylabel("Y=x^2")

plt.show()
```

```python
# 2. Recreate the line graph from the first question, but this time make the line dashed and red in color.


x=np.arange(11)
y=x**2
plt.plot(x,y,":r")
plt.title("Simple LIne Plot")
plt.xlabel("X")
plt.ylabel("Y=x^2")

plt.show()
```

```python
# 3. Modify the plot from the previous question by adding a legend with the label "y = x^2".




x=np.arange(11)
y=x**2
plt.plot(x,y,"--r")
plt.legend(["Y=x^2"])
plt.title("Simple LIne Plot")
plt.xlabel("X")
plt.ylabel("Y=x^2")


plt.show()
```

```python
# 4. Plot two functions on the same graph: y = x^2 and y = x^3. Use different colors for each line and include a legend.


x=np.arange(11)
y1=x**2
y2=x**3
plt.plot(x,y1,"--m",label="Y=x^2")
plt.plot(x,y2,":c",label="Y=x^3")
plt.legend()
plt.title("Simple LIne Plot")
plt.xlabel("X")
plt.ylabel("Y=x^2")


plt.show()
```

```python
# 27. Create a multiple bar plot showing the number of hours spent on three activities (Study, Exercise, Sleep) by three different individuals (Person A, Person B, Person C). Use the data: Person A - [4, 1, 7], Person B - [5, 2, 6], Person C - [3, 1, 8]. Display the bars side by side and use different colors
# for each person.


import matplotlib.pyplot as plt
import numpy as np


Person_A = [4, 1, 7]
Person_B = [5, 2, 6]
Person_C = [3, 1, 8]

activities = ['Study', 'Exercise', 'Sleep']

x = np.arange(len(activities))
width = 0.25

plt.bar(x - width, Person_A, width, label='Person A')
plt.bar(x, Person_B, width, label='Person B')
plt.bar(x + width, Person_C, width, label='Person C')

plt.xlabel('Activities')
plt.ylabel('Hours')
plt.title('Multiple Bar Plot')
plt.xticks(x, activities)
plt.legend()

plt.show()
```

```python
# 28. Create a scatter plot to visualize the relationship between the number
# of hours studied and the scores obtained by 8 students. Use the data: Hours
# studied: [2, 3, 4, 5, 6, 7, 8, 9], Scores: [40, 50, 60, 65, 70, 75, 80, 85]. Label
# the axes and add a title.


import matplotlib.pyplot as plt
import numpy as np

hours_studied = [2, 3, 4, 5, 6, 7, 8, 9]
scores = [40, 50, 60, 65, 70, 75, 80, 85]

plt.scatter(hours_studied, scores)
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Scatter Plot: Hours Studied vs. Scores')

plt.show()
```

```python



# 29. Plot a quadratic function y = -x**2 + 4x for x values from -2 to 4. Use
# fill_between to shade the area under the curve. Make sure to label the axes
# and add a title.


import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 4, 100)
y = -x**2 + 4*x

plt.plot(x, y)
plt.fill_between(x, y, color='blue', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('hi')

plt.show()

```

```python
# 30.Create a 2x2 grid of subplots. In the first subplot (top-left), plot a line graph of y=x**2.
# In the second subplot (top-right), plot a scatter plot of x and y where y is a random sample from a normal distribution. In the third subplot (bottom-left), plot a bar chart of categories A, B, and C with heights 3, 7, and 2, respectively. In the fourth subplot (bottom-right), plot a histogram of 1000 random samples from a normal distribution.



import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = x**2

plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Line Plot')

np.random.seed(0)
y = np.random.randn(100)
plt.subplot(2, 2, 2)
plt.scatter(x, y)
plt.title('Scatter Plot')

categories = ['A', 'B', 'C']
heights = [3, 7, 2]
plt.subplot(2, 2, 3)
plt.bar(categories, heights)
plt.title('Bar Chart')

plt.subplot(2, 2, 4)
samples = np.random.randn(1000)
plt.hist(samples, bins=20)
plt.title('Histogram')

plt.tight_layout()
plt.show()
```
