```python
# 3?


x=[0,1,2,3]
y1=[3,8,1,10]
y2=[6,2,7,11]

plt.plot(x,y1,c="c")
plt.plot(x,y2,c="m")

plt.show()
```

```python
# 4?

y=[3,8,1,10]

plt.plot(y,"o--r",ms=10)

plt.show()
```

```python
# 5?


categories=["A","B","C","D"]
values=[3,7,5,2]

plt.bar(categories,values,color="b")

plt.show()
```

```python
# 6?

categories=["A","B","C","D"]
y1=[10,20,30,40]
y2=[15,25,35,45]


plt.bar(categories,y1)
plt.bar(categories,y2,bottom=y1)

plt.show()
```

```python
# 7?


x=np.random.normal(50,10,1000)

plt.hist(x,bins=20)

plt.show()
```

```python
# 8?


sizes=[25,35,20,40]
labels=["Group A","Group B","Group C","Group D"]
colors=["r","c","m","y"]

plt.pie(sizes,labels=labels,colors=colors)

plt.legend(loc=(1,1))
plt.show()
```

```python
# 9 Draw a pie chart?

sizes=[25,20,45,10]
categories=["A","B","C","D"]
colors=["g","yellow","pink","r"]

plt.pie(sizes,labels=categories,colors=colors,autopct="%1.0f%%")
plt.legend(loc=(1,1))

plt.show()
```

**part-2**

```python
# 11?

x=[1,2,3,4]
y=[10,20,30,40]


plt.plot(x,y,lw=5,alpha=0.5)

plt.show()
```

```python
# 12?

x=[1,2,3,4]
y=[2,4,6,8]

plt.plot(x,y)

for i,j in zip(x,y):
  plt.annotate(f'{i,j}',xy=(i,j),xytext=(i,j+0.3),ha="center",color="k")

plt.show()
```

```python
# 13?

x=[1,2,3,4,5]
y=[10,20,30,40,50]
size=[50,100,200,300,400]
colors=["r","c","yellow","m","pink"]


plt.scatter(x,y,s=size,color=colors)

plt.show()
```

```python
# 14?

categories=["A","B","C","D","E","F"]
a=np.array([3,6,9,4,2,5])
b=np.array([2,8,1,9,7,3])


plt.bar(categories,a,color="r")
plt.bar(categories,-b,color="k")


plt.show()
```

```python
# 15?

values=[35,25,25,15]
catogories=["Apples","Bananas","Cherries","Dates"]
explode=[0.2,0,0,0]

plt.pie(values,labels=catogories,explode=explode,startangle=90,autopct="%1.0f%%")

plt.legend(loc=(1,1))

plt.show()
```
